```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Scatter plot for sepal length vs petal length
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species') # Changed column names to match those in the DataFrame
plt.title('Scatter Plot: sepal_length vs petal_length')
plt.xlabel('sepal_length (cm)')
plt.ylabel('petal_length (cm)')
plt.legend(title='species')
plt.grid(True)
plt.show()
```

```python
# Histogram of petal width
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='petal_width', kde=True, bins=20, color='purple')
plt.title('Histogram of Petal Width')
plt.xlabel('petal_width')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar graph of mean petal length for each species
mean_values = df.groupby('species')['petal_length'].mean()

plt.figure(figsize=(8, 6))
mean_values.plot(kind='bar', color='orange', alpha=0.8)
plt.title('Mean Petal Length per Species')
plt.xlabel('species')
plt.ylabel('Mean Petal Length')
plt.xticks(rotation=0)
plt.show()
```

```python
# Compute correlation matrix
correlation_matrix = df.iloc[:, :-1].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pie chart of species distribution
species_count = df['species'].value_counts()

plt.figure(figsize=(8, 6))
species_count.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'salmon'])
plt.title('Species Distribution')
plt.ylabel('')  # Remove y-axis label
plt.show()
```

```python
# Box plot for petal length grouped by species
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='species', y='petal_length', palette='Set2')
plt.title('Box Plot: Petal Length by Species')
plt.xlabel('species')
plt.ylabel('petal_length')
plt.show()
```
