```python
data = {
    'Year': range(2010, 2022),
    'Internet_Users_Millions': [2023, 2231, 2494, 2710, 2926, 3185, 3424, 3650, 3896, 4131, 4901, 5168],
    'Penetration_Rate': [29.3, 31.8, 35.1, 37.5, 39.9, 42.5, 45.1, 47.6, 50.4, 53.6, 62.5, 65.6],
    'Growth_Rate': [14.5, 10.3, 11.8, 8.7, 8.0, 8.8, 7.5, 6.6, 6.7, 6.0, 18.6, 5.4]
}
df = pd.DataFrame(data)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12))
fig.suptitle('Global Internet Usage Statistics (2010-2021)', fontsize=16, y=0.95)


color1 = plt.cm.Blues(np.linspace(0.6, 0.9, len(df)))
bars1 = ax1.bar(df['Year'], df['Internet_Users_Millions'], color=color1)
ax1.set_title('Number of Internet Users Worldwide')
ax1.set_xlabel('Year')
ax1.set_ylabel('Internet Users (Millions)')
ax1.grid(True, linestyle='--', alpha=0.7)


for bar in bars1:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:,.0f}M',
             ha='center', va='bottom', fontsize=10)


ax1_twin = ax1.twinx()
line = ax1_twin.plot(df['Year'], df['Penetration_Rate'],
                     color='red', marker='o', linewidth=2, label='Penetration Rate')
ax1_twin.set_ylabel('Penetration Rate (%)', color='red')
ax1_twin.tick_params(axis='y', labelcolor='red')


color2 = plt.cm.Greens(np.linspace(0.6, 0.9, len(df)))
bars2 = ax2.bar(df['Year'], df['Growth_Rate'], color=color2)
ax2.set_title('Year-over-Year Growth Rate')
ax2.set_xlabel('Year')
ax2.set_ylabel('Growth Rate (%)')
ax2.grid(True, linestyle='--', alpha=0.7)


for bar in bars2:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.1f}%',
             ha='center', va='bottom', fontsize=10)


stats_text = (f"Summary Statistics (2021):\n"
             f"Total Users: {df['Internet_Users_Millions'].iloc[-1]:,.0f}M\n"
             f"Penetration Rate: {df['Penetration_Rate'].iloc[-1]:.1f}%\n"
             f"Growth Rate: {df['Growth_Rate'].iloc[-1]:.1f}%")

plt.figtext(0.15, 0.02, stats_text, fontsize=10, bbox=dict(facecolor='white', alpha=0.8))

plt.tight_layout()

plt.show()

print("\nDetailed Statistics:")
print(df.describe().round(2))


years = df['Year'].iloc[-1] - df['Year'].iloc[0]
cagr = (((df['Internet_Users_Millions'].iloc[-1] /
          df['Internet_Users_Millions'].iloc[0]) ** (1/years)) - 1) * 100
print(f"\nCompound Annual Growth Rate (CAGR): {cagr:.2f}%")
```

```python


data = {
    'Year': range(2010, 2022),
    'Internet_Users_Millions': [2023, 2231, 2494, 2710, 2926, 3185, 3424, 3650, 3896, 4131, 4901, 5168],
    'Penetration_Rate': [29.3, 31.8, 35.1, 37.5, 39.9, 42.5, 45.1, 47.6, 50.4, 53.6, 62.5, 65.6],
    'Growth_Rate': [14.5, 10.3, 11.8, 8.7, 8.0, 8.8, 7.5, 6.6, 6.7, 6.0, 18.6, 5.4]
}
df = pd.DataFrame(data)

plt.style.use('default')

fig = plt.figure(figsize=(15, 10))
fig.suptitle('Internet Usage Analysis - Scatter Plots', fontsize=16, y=0.95)

ax1 = plt.subplot(221)
scatter1 = ax1.scatter(df['Penetration_Rate'], df['Internet_Users_Millions'],
                      c=df['Year'], cmap='viridis', s=100)
ax1.set_xlabel('Penetration Rate (%)')
ax1.set_ylabel('Internet Users (Millions)')
ax1.set_title('Users vs Penetration Rate')
ax1.grid(True, alpha=0.3)

for i, txt in enumerate(df['Year']):
    ax1.annotate(txt, (df['Penetration_Rate'].iloc[i], df['Internet_Users_Millions'].iloc[i]),
                 xytext=(5, 5), textcoords='offset points', fontsize=8)


plt.colorbar(scatter1, label='Year')


ax2 = plt.subplot(222)
scatter2 = ax2.scatter(df['Growth_Rate'], df['Internet_Users_Millions'],
                      c=df['Year'], cmap='viridis', s=100)
ax2.set_xlabel('Growth Rate (%)')
ax2.set_ylabel('Internet Users (Millions)')
ax2.set_title('Users vs Growth Rate')
ax2.grid(True, alpha=0.3)


for i, txt in enumerate(df['Year']):
    ax2.annotate(txt, (df['Growth_Rate'].iloc[i], df['Internet_Users_Millions'].iloc[i]),
                 xytext=(5, 5), textcoords='offset points', fontsize=8)


plt.colorbar(scatter2, label='Year')


ax3 = plt.subplot(223)
ax3.scatter(df['Penetration_Rate'], df['Growth_Rate'], c=df['Year'],
            cmap='viridis', s=100)
ax3.grid(True, alpha=0.3)

z = np.polyfit(df['Penetration_Rate'], df['Growth_Rate'], 1)
p = np.poly1d(z)
ax3.plot(df['Penetration_Rate'], p(df['Penetration_Rate']), "r--", alpha=0.8,
         label=f'Trend line (y={z[0]:.2f}x{z[1]:+.2f})')

ax3.set_xlabel('Penetration Rate (%)')
ax3.set_ylabel('Growth Rate (%)')
ax3.set_title('Growth Rate vs Penetration Rate')
ax3.legend()

ax4 = plt.subplot(224)
scatter4 = ax4.scatter(df['Year'], df['Growth_Rate'],
                      s=df['Internet_Users_Millions']/30,
                      c=df['Penetration_Rate'], cmap='viridis',
                      alpha=0.6)
ax4.set_xlabel('Year')
ax4.set_ylabel('Growth Rate (%)')
ax4.set_title('Year-wise Growth (size = number of users)')
ax4.grid(True, alpha=0.3)

plt.colorbar(scatter4, label='Penetration Rate (%)')

plt.tight_layout()

correlations = df[['Internet_Users_Millions', 'Penetration_Rate', 'Growth_Rate']].corr()
correlation_text = "Correlations:\n"
correlation_text += f"Users vs Penetration Rate: {correlations.iloc[0,1]:.2f}\n"
correlation_text += f"Users vs Growth Rate: {correlations.iloc[0,2]:.2f}\n"
correlation_text += f"Penetration vs Growth Rate: {correlations.iloc[1,2]:.2f}"

plt.figtext(0.02, 0.02, correlation_text, fontsize=8, bbox=dict(facecolor='white', alpha=0.8))

plt.show()

print("\nDetailed Statistics:")
print(df.describe().round(2))

print("\nYear-over-Year Analysis:")
print(df.set_index('Year').pct_change().dropna().round(3) * 100)
```

```python
data = {
    'Year': range(2010, 2022),
    'Internet_Users_Millions': [2023, 2231, 2494, 2710, 2926, 3185, 3424, 3650, 3896, 4131, 4901, 5168],
    'Penetration_Rate': [29.3, 31.8, 35.1, 37.5, 39.9, 42.5, 45.1, 47.6, 50.4, 53.6, 62.5, 65.6],
    'Growth_Rate': [14.5, 10.3, 11.8, 8.7, 8.0, 8.8, 7.5, 6.6, 6.7, 6.0, 18.6, 5.4]
}
df = pd.DataFrame(data)


correlation_matrix = df.corr()


plt.figure(figsize=(12, 8))

sns.heatmap(correlation_matrix,
            annot=True,
            cmap='coolwarm',
            vmin=-1, vmax=1,
            center=0,
            square=True,
            fmt='.2f',
            linewidths=0.5,
            cbar_kws={"shrink": .8})

plt.title('Correlation Matrix of Internet Usage Metrics', pad=20)


plt.tight_layout()


plt.show()


print("\nDetailed Correlation Analysis:")
print("\nCorrelation Matrix:")
print(correlation_matrix.round(3))


from scipy import stats

def calculate_correlation_significance(df):
    columns = df.columns
    p_values = pd.DataFrame(np.zeros((len(columns), len(columns))),
                           columns=columns, index=columns)

    for i in columns:
        for j in columns:
            _, p_value = stats.pearsonr(df[i], df[j])
            p_values.loc[i,j] = p_value

    return p_values

p_values = calculate_correlation_significance(df)
print("\nP-values (Statistical Significance):")
print(p_values.round(3))


print("\nCorrelation Interpretation:")
for col1 in df.columns:
    for col2 in df.columns:
        if col1 < col2:
            corr = correlation_matrix.loc[col1, col2]
            p_val = p_values.loc[col1, col2]
            strength = ""
            if abs(corr) > 0.7:
                strength = "strong"
            elif abs(corr) > 0.3:
                strength = "moderate"
            else:
                strength = "weak"

            print(f"\n{col1} vs {col2}:")
            print(f"- Correlation coefficient: {corr:.3f}")
            print(f"- Strength: {strength}")
            print(f"- Direction: {'positive' if corr > 0 else 'negative'}")
            print(f"- Statistical significance: {'significant' if p_val < 0.05 else 'not significant'} (p={p_val:.3f})")
```
